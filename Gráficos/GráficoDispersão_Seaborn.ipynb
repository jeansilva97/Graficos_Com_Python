```python
#Importando bibliotecas:
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
#Carregando base de dados:
base = pd.read_csv("co2.csv")
```

```python
#Gerando gráfico de dispersão utilizando os atributos "conc" e "uptake", agrupando pelo "type":
srn.scatterplot(base.conc, base.uptake, hue = base.Type)
```

```python
#Separando e armazendo em variáveis os registros "quebec" e "mississipi" da base de dados:
q = base.loc[base["Type"] == "Quebec"]
m =base.loc[base["Type"] == "Mississippi"]
```

```python
#Gerando no mesmo gráfico dois gráficos de dispersão um com "Quebec" e outro do "Mississipi":
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title("Quebec")
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title("Mississippi")
#Ajeitar o layout:
plt.tight_layout()
```

```python
#Separando e armazendo em variáveis os registros "chilled" e "nochilled" da base de dados:
ch = base.loc[base["Treatment"] == "chilled"]
nc = base.loc[base["Treatment"] == "nonchilled"]
```

```python
#Gerando no mesmo gráfico dois gráficos de dispersão um com "chilled" e outro do "nochilled":
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(ch.conc, ch.uptake).set_title("Chilled")
plt.subplot(1,2,2)
srn.scatterplot(nc.conc, nc.uptake).set_title("Nochilled")
#Ajeitar o layout:
plt.tight_layout()
```

```python
#Carregando base de dados de cancer de esofago:
base2 = pd.read_csv("esoph.csv")
base2.head()
```

```python
#Gráficos entre os atributos "alcgp" = consumo de alcool e "ncontrols" = numero e controle:
srn.catplot(x = "alcgp", y = "ncontrols", data = base2, jitter = False)
```

```python
#Gráfico entre os atributos consumo de alcool e numero de controle, com agrupamento das categorias de consumo de tabacos:
srn.catplot(x = "alcgp", y = "ncontrols", data = base2, col="tobgp")
```
