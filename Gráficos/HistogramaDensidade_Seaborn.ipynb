```python
#Importando bibliotecas:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
#Carregando base de dados:
base=pd.read_csv("trees.csv")
base.head()
```

```python
#Criando um histograma com linha de densidade:
srn.histplot(base.Volume, color="blue", kde=True, stat="density", bins=10).set(title="Volume")
```

```python
#Carregando outra base de dados:
base2 = pd.read_csv("chicken.csv")
base2
```

```python
#Criando agrupamento de todos os alimentos dados aos frangos com o total dos seus pesos:
agrupar = base2.groupby(["feed"])["weight"].sum()
agrupar
```

```python
#Criando gráfico de densidade considerando apenas os pesos dos frango alimentos com "horsebean"
srn.kdeplot(base2.loc[base2["feed"] == "horsebean"].weight).set(title="Horsebean")
```

```python
#Histograma considerando apenas os pesos dos frangos alimentados com "casein":
srn.histplot(base2.loc[base2["feed"] == "casein"].weight, kde=True).set(title="Casein")
```

```python
#Histograma considerando apenas os pesos dos frangos alimentados com "linseed":
srn.histplot(base2.loc[base2["feed"] == "linseed"].weight, kde=True).set(title="Linseed")
```

```python
#Histograma considerando apenas os pesos dos frangos alimentados com "meatmeal":
srn.histplot(base2.loc[base2["feed"] == "meatmeal"].weight, kde=True).set(title="Meatmeal")
```

```python
#Histograma considerando apenas os pesos dos frangos alimentados com "soybean":
srn.histplot(base2.loc[base2["feed"] == "soybean"].weight, kde=True).set(title="Soybean")
```

```python
#Histograma considerando apenas os pesos dos frangos alimentados com "sunflower":
srn.histplot(base2.loc[base2["feed"] == "sunflower"].weight, kde=True).set(title="Sunflower")
```

```python
#Agrupando os seis gráficos com 2 x 3 - 2 linhas e 3 colunas:
plt.figure()
plt.subplot(3,2,1)
srn.histplot(base2.loc[base2["feed"] == "casein"].weight, kde=True).set(title="Casein")
plt.subplot(3,2,2)
srn.histplot(base2.loc[base2["feed"] == "horsebean"].weight, kde=True).set(title="Horsebean")
plt.subplot(3,2,3)
srn.histplot(base2.loc[base2["feed"] == "linseed"].weight, kde=True).set(title="Linseed")
plt.subplot(3,2,4)
srn.histplot(base2.loc[base2["feed"] == "meatmeal"].weight, kde=True).set(title="Meatmeal")
plt.subplot(3,2,5)
srn.histplot(base2.loc[base2["feed"] == "soybean"].weight, kde=True).set(title="Soybean")
plt.subplot(3,2,6)
srn.histplot(base2.loc[base2["feed"] == "sunflower"].weight, kde=True).set(title="Sunflower")
#Ajustando o layout para não haver sobreposição:
plt.tight_layout()


```
